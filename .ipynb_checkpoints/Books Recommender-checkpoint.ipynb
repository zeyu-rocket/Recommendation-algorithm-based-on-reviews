### Books Recommender system using clustering | Collaborative based

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books = pd.read_csv('data/BX-books.csv',sep=";", error_bad_lines=False ,encoding='latin-1')
```

```python
books.head(2)
```

```python
books.shape
```

```python
books.columns
```

```python
books = books [['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher',
       'Image-URL-L']]
```

```python
books.head(2)
```

```python
books.rename(columns={
    "Book-Title": "title",
    "Book-Author": "author",
    "Year-Of-Publication": "year",
    "Publisher": "publisher",
    "Image-URL-L": "img_url"
    
},inplace = True)
```

```python
books.head()
```

```python

users = pd.read_csv('data/BX-Users.csv',sep=";", error_bad_lines=False,encoding='latin-1')
```

```python
users.head()
```

```python
users.shape
```

```python
ratings = pd.read_csv('data/BX-Book-Ratings.csv',sep=";", error_bad_lines=False ,encoding='latin-1')
```

```python
ratings.head(5)
```

```python
print(books.shape)
print(users.shape)
print(ratings.shape)
```

```python
ratings.rename(columns={
    "User-ID": "user_id",
    "Book-Rating": "rating"
},inplace= True) #替换表头名称
```

```python
ratings.head(5)
```

```python
ratings['user_id'].value_counts() #将不同user_id阅读出的图书数量（评分数量）做一个加和
```

```python
ratings.head()
```

```python
x = ratings['user_id'].value_counts() > 200 #筛选出阅读了两百本书以上的用户
```

```python
x[x].shape #这些用户一共有899个
```

```python
y = x[x].index #把这899个的用户id筛选出来
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)] #检查ratings的user_id是否在y中，如果在y中则被保留。所以最后筛选出来的是评价过两百本书以上的用户
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
books.head(2)
```

```python
#找出有ISBN号码的书（筛选掉信息不全的书）
ratings_with_books = ratings.merge(books, on ="ISBN") 
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
#计算每本书有多少评分，并放在num_rating中
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()#计算每本书有多少评分，并放在num_rating中
```

num_rating.head()

```python
num_rating.head()
```

```python
num_rating.rename(columns={"rating": "num_of_rating"}, inplace=True)
```

```python
num_rating.head()
```

```python
ratings_with_books.head()
```

```python
#通过共同的列'title'合并两个数据框，ratings_with_books 和 num_rating
final_rating = ratings_with_books.merge(num_rating, on = 'title' )
```

```python
final_rating.head(2)
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating']>=50 ]
```

```python
final_rating.sample(10)
```

```python
#删除重复数值
final_rating.drop_duplicates(['user_id','title'], inplace = True)
```

```python
final_rating
```

```python
# user_id 透视表成为一列的数值，title 每本书籍将成为透视表的一行， 透视表的值来自rating列 
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
# 有742个用户，888个评分
book_pivot.shape
```

```python
# 用0取代无效数值
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
# 这是一个用户和书本的矩阵，一个用户不可能看过所有书本，所以当书本的评分为0时需要忽略，看中那些用户看过的书本。
# 使用csr matrix压缩矩阵，保留信息 （scipy库）
!pip install scipy
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
# 用于最近邻搜索的类 
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
 # 目的是为了找到用户-书本矩阵中给定用户最相似的几个用户，从而推荐这些相似用户喜欢的书籍。推荐的方式是基于用户之间的评分相似性，即具有相似评分模式的用户会有相似的书籍喜好。这是协同过滤推荐系统的一种实现方式。
 # 237貌似是哈利波特的书
 # 这行代码还不是很明白
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
# 相似书本的距离
distance
```

```python
# 相似书本给出的建议
suggestion
```

```python
# 相似书本的名称
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
# 可以通过这个语句查看列表里序列号对应的书本
book_pivot.index[237]
```

```python
books_name = book_pivot.index
```

```python
# 显示这一列中书本的名称
books_name
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl','wb'))
pickle.dump(books_name, open('artifacts/books_name.pkl','wb'))
pickle.dump(final_rating, open('artifacts/final_rating.pkl','wb'))
pickle.dump(book_pivot, open('artifacts/book_pivot .pkl','wb'))
```
